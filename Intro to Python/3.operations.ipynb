```python
# 1. Aritmetic Operations
# Addition
print( 2 + 3)
print(2.5 + 3.5)
print(2 + 3.5)
#print(2 + '3.5') we cant add int and string
print(True + True)
print(True + False)
print(False + False)
print('Dhruv' + ' Aggarwal' + ' is' + ' a' + ' data science' + ' learner')
print('______________________________________________________________________')
print("")

# Subtraction 
print( 3 - 2)
print(2.5 - 2)
print(2 - 3.5)
print(False - True)
print('______________________________________________________________________')
print("")

#Multiplication
print( 3 * 2)
print(2.5 * 2)
print(2 *3.5)
print(False * True)
print(True * True)
print('______________________________________________________________________')
print("")

#Divison
print( 3 / 2)
print(2.5 /5.0)
print(2.5 / 1)
print(False /True)
print('______________________________________________________________________')
print("")

#Floor Division
print( 3 // 2)
print(2.5 //5.0)
print(2.5 // 1)

print('______________________________________________________________________')
print("")
#Modulo
print( 3 % 2)
print(2.5 %5.0)
print(2.5 % 1)

print('______________________________________________________________________')
print("")

#Exponentiation
print( 3 ** 2)
print(2.5 ** 5.0)
print(2.5 ** 1)
print('______________________________________________________________________')
print("")

#abs -- absolute
print(abs(-35))
print('______________________________________________________________________')
print("")

#pow -- powerprint
print(pow(3,2))
print(pow(5,3))
print('______________________________________________________________________')
print("")
```

```python
#2 Comparission Operator
a = 10
b = 20
# Equal to (==)
print( a == b)  
print("Dhruv" =='Dhruv')
print("Dhruv" == "dhruv")
# Not equal to (!=)
print(a != b)  # Not equal to (!=)
print(a > b)    # Greater than (>)
print(a < b)    # Less than (<)
print(a >= b)  # Greater than or equal to (>=)
print(a <= b)  # Less than or equal to (<=)

```
